## Train Features and Labels

```python
# test features
!ls data/train_features/content/train_features/
```

```python
# test features
!ls data/train_labels/content/train_labels/
```

```python
path_feats = "data/train_features/content/train_features/"
path_labels = "data/train_labels/content/train_labels/"
```

```python
for n in range(20):
    path_feat = datos_train.iloc[n].feature_image
    path_label = datos_train.iloc[n].label_image
    feat = tiff.imread(path_feats + path_feat)
    label = tiff.imread(path_labels + path_label)

    f, axs = plt.subplots(1, 11, figsize=(10, 1))
    
    for i, ax in zip(range(0, 11), axs):
        
        if i < 10:
            ax.imshow(feat[:, :,i])
#             ax.set_axis('off')
            ax.set_xlabel('feat')
        else:
            ax.imshow(label)
#             ax.axis('off')
            ax.set_xlabel('label')

        plt.margins(0)
        
    plt.show()
```
